```python
cols=['radius_mean','concave points_mean','perimeter_mean','area_worst','perimeter_worst']
```

```python
XData=X[cols]
```

```python
XData.shape
```

```python
from sklearn import ensemble
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import joblib
```

```python
import numpy  as np
```

```python
YFinal= np.asarray([1 if c == 'M' else 0 for c in Y])
```

```python
XData.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(XData,YFinal, 
                                                    test_size=0.3,
                                                    random_state=43)
```

```python
print('Shape of  training set: X:{}, y:{}'.format(X_train.shape, y_train.shape))
print('Shape of  test set: X:{}, y:{}'.format(X_test.shape, y_test.shape))
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn import metrics
```

```python
k_range=range(1,26)
```

```python
scores={}
score_list=[]

for k in k_range:
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,y_train)
    y_pred=knn.predict(X_test)
    scores[k]=metrics.accuracy_score(y_test,y_pred)
    score_list.append(metrics.accuracy_score(y_test,y_pred))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(k_range,score_list)
plt.xlabel("Value of k for KNN")
plt.ylabel("Testing Accuracy")
```

```python
knn=KNeighborsClassifier(n_neighbors=9)
knn.fit(X_train,y_train)
```

```python
knn_pred=knn.predict(X_test)
```

```python
print('Accuracy : {}'.format(accuracy_score(y_test, knn_pred)))
```

## SVM
